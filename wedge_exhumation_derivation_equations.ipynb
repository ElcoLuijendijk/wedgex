# Analytical model for the exhumation of an orogenic wedge

## Some definitions:

* L is the width of the wedge
* H is max height of the wedge at right hand side (x=L)
* h is the height of the wedge at any coordinate x
* vx,vy are the horizontal and vertical components of the rock velocity in the wedge
* vc is the compression velocity of the wedge, which is equal to the max. horizontal velocity at the right hand side of the wedge
* vyt is the rock velocity at the top of the wedge
* vym is the maximum rock velocity at the top of the wedge, which is at the right hand side of the wedge (x=L)
* vyb =is the vertical velocity at the base of the wedge
* b is the elevation of the bottom of the wedge above a reference point
* Alpha is the slope of the land surface (m/m)
* beta is the slope of the bottom of the wedge (m/m)

## Some boundary conditions & assumptions

* The vertical velocity of the rocks increases linearly from the left-hand side to the right hand side
* The vertical velocity at the bottom of the wedge (y=b) equals zero
* The horizontal velocity increases linearly from vx=0 at the left hand side of the wedge (x=0) to vx=vxc at the right hand side of the wedge (x=L)
* Assume the length of L stays constant, ie material is constantly accreted to the edge of the wedge and the distribution of horizontal and vertical velocities stays constant. This is maybe the best option



## Derivation vertical velocities in wedge as a result of uniform compression

For compression without any volume loss the follow volume balance can be written as:

$$v_{c} H = \int_0^L{v_{yt}} L$$

The velocity at the top of the wedge is assumed to increase linearly from vyt=0 at the left hand side to vyt=vym at the right hand side (x=L):

$$v_{yt} = \dfrac{x}{L} v_{ym}$$

The area that the wedge extends at the top should be equal to the area that is compressed at the right hand side:

$$\frac{1}{2} L v_{ym} = -H v_c$$ 

which can be combined with the previous eq. to form:

$$\dfrac{L}{x} v_{yt} = -2 \dfrac{H}{L} v_c$$

which can be reorganized as:

$$v_{yt} = -2 \dfrac{H}{L^2} x v_c$$


if we simplify and assume that the vertical velocity at the bottom of the wedge is zero then:

$$v_y = \dfrac{(y-b)}{h} v_{yt}$$


this can be rewritten to yield and expression for v_yt

$$v_{yt} = \dfrac{h}{(y-b)} v_y$$

inserting this into the previous equation yields an expression for vy:

$$\dfrac{h}{(y-b)} v_y = -2 \dfrac{H}{L^2} x v_c$$

which after reorganzing becomes:

$$v_y = -2 \dfrac{H}{L^2} \dfrac{(y-b)}{h} x v_c$$

This can be further simplified by using the following expression for the height of the wedge:

$$h = \dfrac{x}{L} H$$

which after inserting this into the equation yields

$$v_y = -2 \dfrac{H}{L^2} \dfrac{(y-b) L}{H x} x v_c$$

which can be simplified to

$$v_y = -\dfrac{2 (y-b)}{L} v_c$$

rewriting $b = \beta x$

$$v_y = -\dfrac{2 (y- (\beta x))}{L} v_c$$

## Integrating horizontal velocity:

First we'll develop and equation for horizontal velcotiy and intergate this, which gives an equation of the x coordinate of a rock particle versus time.

Horizontal velcoty is assumed to increase linearly from 0 at the left-hand side to vx=vc at the right hand side:

$$\dfrac{\partial x}{\partial t} = \dfrac{x}{L}v_c$$

which can be rewritten as:

$$\dfrac{\partial x}{\partial t} = \dfrac{x}{L}v_c$$

rearringing gives

$$\dfrac{1}{x} \partial x = \dfrac{v_c}{L} \partial t$$

which can be integrated:
$$\int{\dfrac{1}{x}} \partial x = \int{\dfrac{v_c}{L}} \partial t$$

to yield:

$$\ln |x| = \dfrac{v_c t}{L} + C$$

$$x = C_2 e^{v_c t /L}$$

for t=0 x=x0, which yields C2=x0

$$x = x_0 e^{v_c t /L}$$

## Integration to get a function of y vs time:

Next we will integrate the equation for vertical velocity to get a function of the vertical position of a rock particel (y) over time.

note that $v_y = \dfrac{\partial y}{\partial t}$

which can be inserted in the expression for vy:

$$\dfrac{\partial y}{\partial t} = -\dfrac{2 (y-b)}{L} v_c$$

rearringing gives

$$\dfrac{1}{(y-b)} \partial y = \dfrac{-2 v_c}{L} \partial t$$

We add a simplification and assume that b is constant. Note that as a result this means that this equation can only be used for short timescales, at which b (the base of the wedge for each rock particle) does not change too much.

See next box for an integration without this assumption using separation of variables.

With this simplification intergrating both sides yields:

$$\int{\dfrac{1}{(y-b)}} \partial y = \int{\dfrac{-2 v_c}{L}} \partial t$$

which yields:
$$ln(y-b) = \dfrac{-2 v_c}{L} t + C$$

where C is a constant

$$y = e^C e^{-2 v_c t /L} + b$$

defining $C_1 = e^C$ yields:

$$y = C_1 e^{-2 v_c t /L} + b$$

at $t=0$ $y=y_t$ and therefore:

$$C1 = y_t - b = h$$

which leads to

$$y = h e^{-2 v_c t /L} + b$$

with $b = \beta x$ and $h = (\alpha - \beta) x$ this yields:

$$y = ((\alpha - \beta) x) e^{-2 v_c t /L} + \beta x$$

$$\dfrac{\partial y}{\partial t} = 2c (y - \beta x)$$

## New try with separation of variables

The correct form of the vertical velocity eq. is:
$$v_y = \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) y + (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta}) x \right)$$

replacing $v_y = \dfrac{\partial y}{\partial t}$

$$\dfrac{\partial y}{\partial t} = \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) y + (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta}) x \right)$$

This can be simplified as

$$\dfrac{\partial y}{\partial t} = C_1 y + C_2 x $$

where 

$$C_1 =  \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) \right)$$

and
$$C_2 = \dfrac{v_c}{L} (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta})$$

adding the result of the integration of vx over time:


$$\dfrac{\partial y}{\partial t} = C_1 y + C_3 e^{C4 t} $$

where 
$$C_3 = C_2 x_0$$

and 

$$C_4 = v_c / L$$

rewriting this to a form for which a standard solution exists:

$$\partial y - C_1 y =  C_3 e^{C4 t} \partial t$$

The solution for this equation is:

$$y = \dfrac{\int{\mu q(t) \partial t}}{\mu}$$

where $\mu$ is a trail function and $q(t)$ is the right hand side term $q(t) = C_3 e^{C4 t} \partial t$

the solution of the trail function is:

$$\mu = e^{\int{p(y)} \partial y}$$

where $p(y)$ is the coefficient before the y term:

$$p(y) = -C_1$$

and therefore:

$$\mu = e^{\int -C_1 \partial t} = e^{-C_1 t}$$

inserting this back into the solution yields:


$$y = \dfrac{\int{e^{-C_1 t} C_3 e^{C_4 t}} \partial t}{e^{-C_1 t}}$$

or 

$$y = \dfrac{\int{C_3 e^{C_4-C_1 t} } \partial t}{e^{-C_1 t}}$$


which can be integrated as


$$y =  \dfrac{\dfrac{C_3}{C_4-C_1} e^{(C4-C1)t}}{e^{-C1 t}}$$

which can be simplified to:

$$y =  \dfrac{C_3}{C_4-C_1} e^{C4t}$$

reinserting the constants:

$$y =  \dfrac{x_0 C_2}{C_4-C_1} e^{(v_c/L) t}$$

and some more:

$$y =  \dfrac{x_0 C_2}{C_4-C_1} e^{(v_c/L) t}$$

which can be further simplified to:

$$y =  - \dfrac{2 \beta + \dfrac{\alpha \beta}{\alpha - \beta}}{2 + \dfrac{\beta}{(\alpha - \beta)}} x_0 e^{(v_c/L) t}$$

and even further to

$$y = - \dfrac{\beta (3 \alpha - 2 \beta)}{2 \alpha - \beta} x_0 e^{(v_c/L) t}$$
