
$$C_5 =  \left( (-2 - \dfrac{\beta}{\alpha - \beta}) \right)$$

$$C_6 = (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta})$$


The term $\dfrac{C_6}{1 - C_5}$ can be closely approximated by $\beta$, see numerical proof below:

```python
import numpy as np
import itertools
import matplotlib.pyplot as pl

def test_1st_term_eq(alpha, beta):
    """
    check if we can simplify C6 / (1 - C5)
    """

    C5 = -2 - (beta / (alpha - beta))
    C6 = 2 * beta + (alpha * beta) / (alpha - beta)
    #c = 2*b + (a*b)/(a-b) / (3 + b/(a-b))
    
    return (C6 / (1- C5))

a = np.arange(0.005, 0.1, 0.005)
b = -np.arange(0.005, 0.1, 0.005)

combs = list(itertools.product(a, b))

a_vals = np.array([c[0] for c in combs])
b_vals =  np.array([c[1] for c in combs])

cs = test_1st_term_eq(a_vals, b_vals)

## figure:
fig, axs = pl.subplots(1, 2, figsize=(8, 4))

la = axs[0].scatter(a_vals, b_vals, c=cs)
axs[0].set_xlabel(r'$\alpha$')
axs[0].set_ylabel(r'$\beta$')

cb = fig.colorbar(la, ax=axs[0])
cb.set_label('value C6/ (1-C5)')


lb = axs[1].scatter(b_vals, cs)

axs[1].set_xlabel(r'$\beta$')
axs[1].set_ylabel('value C6/ (1-C5)')

fig.tight_layout()
```
