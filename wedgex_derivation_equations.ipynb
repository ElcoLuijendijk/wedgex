# Analytical model for the exhumation of an orogenic wedge

Below is a conceptual model of an orogenic wedge that is simulatanously under compression and transported over a basal detachment:

![Conceptual model for an orogenic wedge that moves up a sloping detachment and is compressed, with the surface staying at the same level](fig/Wedge_Model_Conceptual_smaller.png)

## Some definitions:

* L is the width of the wedge
* $v_d$ is the velcoity of the wedge itself
* $v_c$ is the compression velocity of the wedge, which is equal to the max. horizontal velocity at the right hand side of the wedge
* $\alpha$ is the slope of the land surface (m/m)
* $\beta$ is the slope of the bottom of the wedge (m/m)
* $v_x$, $v_y$ are the horizontal and vertical components of the rock velocity in the wedge
* H is max height of the wedge at right hand side (x=L)
* h is the height of the wedge at any coordinate x
* $v_{yt}$ is the rock velocity at the top of the wedge
* $v_{ym}$ is the maximum rock velocity at the top of the wedge, which is at the right hand side of the wedge (x=L)
* $v_{yb}$ is the vertical velocity at the base of the wedge
* b is the elevation of the bottom of the wedge above a reference point

## Some boundary conditions & assumptions

* The system is in steady-state: The wedge geometry and the land surface stay at the same position, and as much material is added to the right-hand side of the wedge as is removed from the top of the wedge.
* The vertical velocity of the rocks increases linearly from the left-hand side to the right hand side
* Rock particles at the bottom of the wedge move parallel to the bottom of the wedge
* The horizontal velocity increases linearly from $v_x=0$ at the left hand side of the wedge ($x=0$) to $v_x=v_c$ at the right hand side of the wedge ($x=L$)
* The length of the wedge (L) stays constant, i.e., material is constantly accreted to the edge of the wedge and the distribution of horizontal and vertical velocities within the wedge are constant

## Derivation vertical velocities in wedge as a result of uniform compression

For compression without any volume loss the follow volume balance can be written as:

$$v_{c} H = \int_0^L{v_{yt}} L$$

The velocity at the top of the wedge is assumed to increase linearly from vyt=0 at the left hand side to vyt=vym at the right hand side (x=L):

$$v_{yt} = \dfrac{x}{L} v_{ym}$$

The area that the wedge extends at the top should be equal to the area that is compressed at the right hand side:

$$\frac{1}{2} L v_{ym} = -H v_c$$ 

which can be combined with the previous eq. to form:

$$\dfrac{L}{x} v_{yt} = -2 \dfrac{H}{L} v_c$$

which can be reorganized as:

$$v_{yt} = -2 \dfrac{H}{L^2} x v_c$$

## Solution with simplification vy=0 at bottom of the wedge


We simplify and assume that the vertical velocity at the bottom of the wedge is zero then:

$$v_y = \dfrac{(y-b)}{h} v_{yt}$$


this can be rewritten to yield and expression for v_yt

$$v_{yt} = \dfrac{h}{(y-b)} v_y$$

inserting this into the previous equation yields an expression for vy:

$$\dfrac{h}{(y-b)} v_y = -2 \dfrac{H}{L^2} x v_c$$

which after reorganzing becomes:

$$v_y = -2 \dfrac{H}{L^2} \dfrac{(y-b)}{h} x v_c$$

This can be further simplified by using the following expression for the height of the wedge:

$$h = \dfrac{x}{L} H$$

which after inserting this into the equation yields

$$v_y = -2 \dfrac{H}{L^2} \dfrac{(y-b) L}{H x} x v_c$$

which can be simplified to

$$v_y = -\dfrac{2 (y-b)}{L} v_c$$

rewriting $b = \beta x$

$$v_y = -\dfrac{2 (y- (\beta x))}{L} v_c$$

## Solution with correct velocity at the bottom of the wedge:

Instead of assuming a zero vertical velocity at the bottom of the wedge, we assume that the vertical velocity at any point (x,y) increases linearly from v_b at the bottom of the wedge (y=b) to vyt at the top of the wedge:

$$v_y = f v_{y_b} + (1-f) v_{y_t}$$

$$v_yt = \dfrac{(v_y - (1-f) v_{yb})}{f}$$

where $f$ is the relative position of the rock particle: $(y-b)/h$

The original eq. for vyt that was derived previously is:

$$v_{yt} = -2 \dfrac{H}{L^2} x v_c$$

combining these two equations yields:

$$\dfrac{(v_y - (1-f) v_{yb})}{f} = -2 \dfrac{H}{L^2} x v_c$$

which after reorganizing becomes:

$$ v_y = -2 \dfrac{H}{L^2} x v_c f + (1-f) v_{yb}$$

reinserting the original expression for f = (y-b)/h results in:

$$ v_y = -2 \dfrac{H}{L^2} x v_c (y-b)/ h + (1-(y - b)/h) v_{yb}$$

and rewriting $b = \beta x$ and $h = (\alpha - \beta) x = \gamma x$

$$ v_y = \dfrac{- 2 H x v_c}{L^2}  \dfrac{y - \beta x}{\gamma x} + \dfrac{\alpha x - y}{\gamma x} \alpha v_c x / L$$

with $H = (\alpha - \beta) L = \gamma L$:

$$v_y = \dfrac{- 2 v_c (y - \beta x)}{L}   + \dfrac{(\alpha x - y) \beta v_c }{\gamma L} $$

rearranging:

$$v_y = \dfrac{v_c}{L} \left[ \left( -2y + 2 \beta x \right) + \left( \dfrac{\alpha \beta}{\gamma} x - \dfrac{\beta}{\gamma} y \right) \right]$$

rearranging some more:

$$v_y = \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\gamma}) y + (2 \beta + \dfrac{\alpha \beta}{\gamma}) x \right)$$

and even more:

$$v_y = \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) y + (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta}) x \right)$$


**Note this equation is correct, see checks in model_test notebook**

## Integrating horizontal velocity:

First we'll develop and equation for horizontal velcotiy and intergate this, which gives an equation of the x coordinate of a rock particle versus time.

Horizontal velocity is assumed to increase linearly from 0 at the left-hand side to $v_x=v_c$ at the right hand side:

$$\dfrac{\partial x}{\partial t} = \dfrac{x}{L}v_c$$

rearringing gives

$$\dfrac{1}{x} \partial x = \dfrac{v_c}{L} \partial t$$

which can be integrated:
$$\int{\dfrac{1}{x}} \partial x = \int{\dfrac{v_c}{L}} \partial t$$

to yield:

$$\ln |x| = \dfrac{v_c t}{L} + C$$

which can be written as:

$$x = C_2 e^{v_c t /L}$$

for $t=0$ $x=x_0$, which yields $C_2=x_0$

$$x = x_0 e^{v_c t /L}$$

## Integration of vertical velocity to get a function of y vs time:

Next we will integrate the equation for vertical velocity to get a function of the vertical position of a rock particle (y) over time.

note that $v_y = \dfrac{\partial y}{\partial t}$

which can be inserted in the expression for vy derived in the previous section. Note that this is the simplified eq. for $v_y$ that assumes $v_y$ is zero at the bottom of the wedge (which is not 100% correct):

$$\dfrac{\partial y}{\partial t} = -\dfrac{2 (y-b)}{L} v_c$$

rearringing gives

$$\dfrac{1}{(y-b)} \partial y = \dfrac{-2 v_c}{L} \partial t$$

We add a simplification and assume that b is constant. Note that as a result this means that this equation can only be used for short timescales, at which b (the elevation of the base of the wedge below each rock particle) does not change too much.

See next box for an integration without this assumption using separation of variables.

With this simplification intergrating both sides yields:

$$\int{\dfrac{1}{(y-b)}} \partial y = \int{\dfrac{-2 v_c}{L}} \partial t$$

which yields:
$$ln(y-b) = \dfrac{-2 v_c}{L} t + C$$

where C is a constant

$$y = e^C e^{-2 v_c t /L} + b$$

defining $C_1 = e^C$ yields:

$$y = C_1 e^{-2 v_c t /L} + b$$

at $t=0$ $y=y_t$ and therefore:

$$C_1 = y_t - b = h$$

which leads to

$$y = h e^{-2 v_c t /L} + b$$

with $b = \beta x$ and $h = (\alpha - \beta) x$ this yields:

$$y = ((\alpha - \beta) x) e^{-2 v_c t /L} + \beta x$$

## Integration of vertical velocity with separation of variables

The previous integration is not 100% correct, because x is in itself a function of t and can therefore not be regarded as a constant. We therefore need to rewrite x as a function of t and try integrating this equation instead.

$$\dfrac{\partial y}{\partial t} = -\dfrac{2 (y-b)}{L} v_c$$

rewriting as

$$\dfrac{\partial y}{\partial t} = -\dfrac{2 v_c}{L} (y-b)$$

$$\dfrac{\partial y}{\partial t} = 2c (y-b)$$

where $c = -v_c / L$

Rewriting b as a function of x:

$$\dfrac{\partial y}{\partial t} = 2c (y - \beta x)$$

and inserting the solution for x over time:

$$\dfrac{\partial y}{\partial t} = 2c (y - \beta x_0 e^{c t})$$

$$\dfrac{\partial y}{\partial t} = 2c y - 2c \beta x_0 e^{c t}$$

$$\dfrac{\partial y}{\partial t} = 2c y - l e^{c t}$$

and $l=2c \beta x_0$

solution (see notes):

$$y = \dfrac{l}{c} e^{c t} + C e^{2c t}$$

with $y=y_t$ at $t=0$ this yields:

$$C=y_t - \dfrac{l}{c}$$

inserting this back into the equation yields:

$$y = \dfrac{l}{c} e^{c t} + (y_t - 1) e^{2c t}$$

substituting in back the variables and subsituting $y_t = \alpha x_0$ results in this equation:

$$y = \dfrac{-2 v_c /L \beta x_0 L}{v_c} e^{-v_c t / L} + (y_t - 2 \beta x_0) e^{-2 v_c t / L}$$

which can be reorganized to:

$$y = 2 \beta x_0 e^{-v_c t / L} + (\alpha x_0 - 2 \beta x_0) e^{-2 v_c t / L}$$

## Solution with the correct form for vertical velocity:

The previous result was derived for the simplified form of the vertical velocity. The correct form of the vertical velocity eq. is:

$$v_y = \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) y + (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta}) x \right)$$

replacing $v_y = \dfrac{\partial y}{\partial t}$

$$\dfrac{\partial y}{\partial t} = \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) y + (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta}) x \right)$$

This can be simplified as

$$\dfrac{\partial y}{\partial t} = C_1 y + C_2 x $$

where 

$$C_1 =  \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) \right)$$

and
$$C_2 = \dfrac{v_c}{L} (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta})$$

The solution, as derived by Leo is:

$$y=C_2 x_0 \dfrac{e^{v_c t /L}}{v_c/L - C_1} + \left( y_0 - \dfrac{C_2 x_0}{v_c /L - C_1} \right) e ^{C_1 t}$$

## Simplifying the solution

This can be simplified further. Defining v as the normalized compression velocity 

$$v = v_c / L$$

and redefining the constants $C_1$ and $C_2$:

$$C_1 =  v \left( (-2 - \dfrac{\beta}{\alpha - \beta}) \right) = v C_5$$

$$C_2 = v (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta}) = v C_6$$

results in:

$$y= \dfrac{C_6 v}{v - C_5 v} x_0 e^{v t} + \left( y_0 - \dfrac{C_6 v}{v - C_5 v} x_0 \right) e ^{C_5 v t}$$

which simplies to:

$$y= \dfrac{C_6}{1 - C_5} x_0 e^{v t} + \left( y_0 - \dfrac{C_6}{1 - C_5} x_0 \right) e ^{C_5 v t}$$


The term $\dfrac{C_6}{1 - C_5}$ equals $\beta$ (see empirical proof in notebook [proof_for_simplifying_C5_C6_term.ipynb](proof_for_simplifying_C5_C6_term.ipynb))

which allows even more simplification:

$$y= \beta x_0 e^{v t} + \left( y_0 - \beta x_0 \right) e^{C_5 v t}$$

and taking $y_0 = \alpha x_0$ results in:

$$y= \beta x_0 e^{v t} + \left(\alpha - \beta \right) x_0 e^{C_5 v t}$$


## Solutions that include the transport of the wedge

The previous solutions describe the displacement of particles inside the wedge as a result of compression of the wedge. 

Here we include an additional motion of particles in the wedge by the transport (i.e. sliding) of the wedge along a basal detachment (left and up in the figure).

If we include transport of the wedge along the basal detachment the additional horizontal velocity is:

$$v_{xd} = v_d$$

where $v_d$ is defined as the horizontal velocity of the wedge. Note that this is a bit different than shown in the figure, where $v_d$ is the velocity parallel to the detachment. Using horizontal transport velocity is a bit simpler.

The vertical transport velocity is the product of the slope of the basal detachment and the horizontal transport velocity:

$$v_{yd} = \beta v_d$$


### Horizontal velocity:

The horizontal velocity is the combination of the compression and transport velocities:

$$v_x = v_{xc} + v_d = v x + v_d$$

where $$v = v_c / L$$

Rewriting to a differential equation:

$$\dfrac{\partial x}{\partial t} = v x + v_d$$

integration yields:

$$x = C e^{v t} - \dfrac{v_d}{v}$$

taking $x=x_0$ at $t=0$ yields:

$$x = \left( x_0 + \dfrac{v_d}{v} \right) e^{v t} - \dfrac{v_d}{v}$$


### Vertical velocity

The vertical velcotiy is the sum of the velocity due to compression as derived previously and the transport velocity:

$$v_y = v_{yc} + v_{yd}$$

with the previously derived velocity due to compression:

$$v_{yc} = C_1 y + C_2 x $$

where 

$$C_1 =  \dfrac{v_c}{L} \left( (-2 - \dfrac{\beta}{\alpha - \beta}) \right)$$

and
$$C_2 = \dfrac{v_c}{L} (2 \beta + \dfrac{\alpha \beta}{\alpha - \beta})$$

The combined equation for vertical velocity due to compression and transport is:

$$v_y = C_1 y + C_2 x + \beta v_d$$

Inserting the expression for x derived above yields:

$$v_y = C_1 y + C_2 \left( \left( x_0 - \dfrac{v_d}{v} \right) e^{v t} - \dfrac{v_d}{v} \right) + \beta v_d$$

which can be shortened to

$$\dfrac{\partial y}{\partial t} = C_1 y + C_7 e^{v t} + C_8$$

where

$$C_7 = C_2 (x_0 - \dfrac{v_d}{v})$$

and

$$C_8 = \beta v_d - C_2 \dfrac{v_d}{v}$$

which is hopefully still solvable with boundary condition $y=y_0$ at $t=0$ ....
